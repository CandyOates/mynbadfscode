```python
import numpy as np
import pandas as pd
import datetime

import player as data_pl
import team as data_te
import game as data_ga
```

#### Get Game Logs for all Teams

```python
tid2team = dict(zip([int(data_te.TEAMS[team]['id']) for team in data_te.TEAMS.keys()], data_te.TEAMS.keys()))

table = pd.DataFrame()
for team, dic in data_te.TEAMS.iteritems():
    tid = dic['id']
    gl = data_te.TeamGameLogs(tid, season='2015-16').info()
    table = table.append(gl, ignore_index=True)
table['GAME_DATE'] = [datetime.datetime.strptime(x, '%b %d, %Y').date() for x in table.GAME_DATE]
table
```

## This Script iteratively does what the several following cells does

```python
iDate = datetime.date(2015,10,27)
while iDate <= datetime.datetime.now().date():
    sub_table = table[table.GAME_DATE==iDate]
    if len(sub_table) == 0:
        print iDate, ' has no games'
        bayesian_model = pd.read_csv('ReboundingParameters/team_reb_%s' % iDate).set_index('Team_ID')
        bayesian_model.to_csv('ReboundingParameters/team_reb_%s' % (iDate + datetime.timedelta(days=1)))
        iDate += datetime.timedelta(days=1)
        continue
    
    dic = {'Team_ID': [], 'OREB': [], 'DREB': [], 'NREB': [], 'OPP_OREB': [], 'OPP_DREB': [], 'OMISS': [], 'DMISS': [], 'MIN': [], 'POS': []}
    for tid in np.unique(sub_table.Team_ID):
        gids = np.unique(sub_table[table.Team_ID==tid].Game_ID)
        for gid in gids:
            sub = sub_table[(sub_table.Team_ID==tid) & (sub_table.Game_ID==gid)]
            omiss = np.sum(sub.FGA - sub.FGM + .44*(sub.FTA - sub.FTM))
            oreb = np.sum(sub.OREB)
            dreb = np.sum(sub.DREB)
            subo = sub_table[(sub_table.Team_ID!=tid) & (sub_table.Game_ID==gid)]
            dmiss = np.sum(subo.FGA - subo.FGM + .44*(subo.FTA - subo.FTM))
            opp_oreb = np.sum(subo.OREB)
            opp_dreb = np.sum(subo.DREB)
            nreb = dmiss - opp_oreb - dreb

            mins = np.sum(subo.MIN)
            pos = 0.5 * ((sub.FGA.sum() + 0.4 * sub.FTA.sum() - 1.07 * (sub.OREB.sum() / (sub.OREB.sum() + subo.DREB.sum())) * (sub.FGA.sum() - sub.FGM.sum()) + sub.TOV.sum()) +\
                         (sub.FGA.sum() + 0.4 * subo.FTA.sum() - 1.07 * (subo.OREB.sum() / (subo.OREB.sum() + sub.DREB.sum())) * (subo.FGA.sum() - subo.FGM.sum()) + subo.TOV.sum()))

            dic['Team_ID'].append(tid)
            dic['OREB'].append(oreb)
            dic['DREB'].append(dreb)
            dic['NREB'].append(nreb)
            dic['OPP_OREB'].append(opp_oreb)
            dic['OPP_DREB'].append(opp_dreb)
            dic['OMISS'].append(omiss)
            dic['DMISS'].append(dmiss)
            dic['MIN'].append(mins)
            dic['POS'].append(pos)
    table2 = pd.DataFrame(dic)
    table2 = table2.set_index('Team_ID')
    
    t = pd.read_csv('ReboundingParameters/team_reb_%s' % iDate).set_index('Team_ID')
    t.loc[table2.index] = t.loc[table2.index] + table2[['OREB','OMISS','DREB','DMISS','OPP_OREB','OPP_DREB']]
    t.to_csv('ReboundingParameters/team_reb_%s' % (iDate + datetime.timedelta(days=1)))

    print iDate
    iDate += datetime.timedelta(days=1)
```

#### Strip Out Rebounding Statistics

```python
dic = {'Team_ID': [], 'OREB': [], 'DREB': [], 'NREB': [], 'OPP_OREB': [], 'OPP_DREB': [], 'OMISS': [], 'DMISS': [], 'MIN': [], 'POS': []}
for tid in np.unique(table.Team_ID):
    gids = np.unique(table[table.Team_ID==tid].Game_ID)
    for gid in gids:
        sub = table[(table.Team_ID==tid) & (table.Game_ID==gid)]
        omiss = np.sum(sub.FGA - sub.FGM + .44*(sub.FTA - sub.FTM))
        oreb = np.sum(sub.OREB)
        dreb = np.sum(sub.DREB)
        subo = table[(table.Team_ID!=tid) & (table.Game_ID==gid)]
        dmiss = np.sum(subo.FGA - subo.FGM + .44*(subo.FTA - subo.FTM))
        opp_oreb = np.sum(subo.OREB)
        opp_dreb = np.sum(subo.DREB)
        nreb = dmiss - opp_oreb - dreb
        
        mins = np.sum(subo.MIN)
        pos = 0.5 * ((sub.FGA.sum() + 0.4 * sub.FTA.sum() - 1.07 * (sub.OREB.sum() / (sub.OREB.sum() + subo.DREB.sum())) * (sub.FGA.sum() - sub.FGM.sum()) + sub.TOV.sum()) +\
                     (sub.FGA.sum() + 0.4 * subo.FTA.sum() - 1.07 * (subo.OREB.sum() / (subo.OREB.sum() + sub.DREB.sum())) * (subo.FGA.sum() - subo.FGM.sum()) + subo.TOV.sum()))
        
        dic['Team_ID'].append(tid)
        dic['OREB'].append(oreb)
        dic['DREB'].append(dreb)
        dic['NREB'].append(nreb)
        dic['OPP_OREB'].append(opp_oreb)
        dic['OPP_DREB'].append(opp_dreb)
        dic['OMISS'].append(omiss)
        dic['DMISS'].append(dmiss)
        dic['MIN'].append(mins)
        dic['POS'].append(pos)
table2 = pd.DataFrame(dic)
table2
```

#### Process Rebounding Statistics into New Statistics

```python
pt = pd.pivot_table(pd.DataFrame(dic), values=['DMISS','DREB','NREB','OMISS','OREB','OPP_DREB','OPP_OREB','MIN','POS'], index='Team_ID', aggfunc=np.sum)
pt['DREB%'] = pt.DREB / pt.DMISS
pt['OREB%'] = pt.OREB / pt.OMISS
pt['OPP_DREB%'] = pt.OPP_DREB / pt.OMISS
pt['OPP_OREB%'] = pt.OPP_OREB / pt.DMISS
pt['NREB%'] = pt.NREB / pt.DMISS
pt['MISS'] = pt.OMISS + pt.DMISS
pt['OPP_REB'] = pt.OPP_OREB + pt.OPP_DREB
pt['OPP_REB%'] = pt.OPP_REB / pt.MISS
pt['REB'] = pt.OREB + pt.DREB
pt['REB%'] = pt.REB / pt.MISS
pt['Team'] = [tid2team[tid] for tid in pt.index]
pt['OREB/MIN'] = pt.OREB / pt.MIN
pt['DREB/MIN'] = pt.DREB / pt.MIN
pt['REB/MIN'] = (pt.OREB + pt.DREB) / pt.MIN
pt['POS/MIN'] = pt.POS / pt.MIN
pt.sort('REB%', ascending=False)[['Team','REB%','DREB%','OREB%','OPP_DREB%','OPP_OREB%']]
```

#### Update Bayesian Model

```python
bayesian_model = pt[['OREB','OMISS','DREB','DMISS','OPP_OREB','OPP_DREB']].copy()
bayesian_model[['OREB','OMISS','DREB','DMISS','OPP_OREB','OPP_DREB']] *= (5. / 82)
bayesian_model.to_csv('ReboundingParameters/team_reb_%s' % datetime.date(2015,10,27))
```
