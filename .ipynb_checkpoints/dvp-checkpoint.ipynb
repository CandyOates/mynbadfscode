```python
def reg_non_param(x, bdwidth, x_sample, y_sample):
    """Values of the non-parametric regression of Y wrt X using a Gaussian kernel.

    Parameters
    ----------
    x: numpy array, one dimensional
        Values at which the regression is evaluated
    bdwidth: positive float, value of the bandwidth parameter
    x_sample: numpy array, one dimensional, non-empty
        x values of the sample
    y_sample: numpy array, one dimensional
        y values of the sample, must have the same length as x_sample.    
    """
    def kern(u, x):
        """Gaussian kernel function"""
        return np.exp(-(u[:, np.newaxis] - x) ** 2 / (2 * bdwidth ** 2))

    tmp = kern(x_sample, x)
    return np.sum(tmp * y_sample[:, np.newaxis], axis=0) \
        / np.sum(tmp, axis=0)

def basis(knots, x):
    """Values of order-1 B-spline basis functions.
    
    For an increasingly sorted collection of knots and a collection of
    query points x, returns a 2-dimensional array of values, of dimension
    len(x) x len(knots).
    
    Parameters
    ----------
    knots: numpy array, one dimensional, increasingly sorted
        Knots of the B-spline function
    x: numpy array, one dimensional
        Query points where to evaluate the basis functions.
    """
    nb_knots = len(knots)
    diag = np.identity(nb_knots)
    res = np.empty((len(x), nb_knots))
    for i in xrange(nb_knots):
        res[:, i] = np.interp(x, knots, diag[i])
    return res

def reg_param_coeffs(knots, x_sample, y_sample):
    """Computes the coefficients of the P-L regression of y_sample wrt. x_sample.
    
    For an increasingly sorted collection of knots and two one-dimensional
    samples x_sample and y_sample, computes the PL-regression.
    
    Parameters
    ----------
    knots: numpy array, one dimensional, increasingly sorted
        Knots of the B-spline function
    x_sample: numpy array, one dimensional
        X sample
    y_sample: numpy array, one dimensional
        Y sample
        
    Notes
    -----
    The length of the X and Y samples must be the same.
    The length of the knots array should be at least one.
    The knots must be increasingly sorted.
    """
    bis = basis(knots, x_sample)
    var = bis.T.dot(bis)
    covar = y_sample.dot(bis)
    return np.linalg.lstsq(var, covar.T)[0]

def eval_piecewise_linear(x, knots, coeffs):
    """Eveluates the piecewise linear function at the specified x for the
    specified knots and coeffs.
    
    This is simply a wrapper around np.interp.
    """
    return np.interp(x, knots, coeffs)

def design(x):
    n = len(x)
    return np.array([
#             np.ones(n),
            x
        ]).T

def reg_lin(x, y, wgt=None):
    A = design(x)
    b = y
    n = len(y)
    if wgt is None:
        W = np.eye(n)
    else:
        W = sparse.spdiags(wgt, 0, n, n, format='csc')
    return np.linalg.solve(A.T.dot(W.dot(A)), A.T.dot(W.dot(b)))

def reg_eval(X, coef):
    A = design(X)
    return A.dot(coef)
```

# Model Each Statistic Against Each Team by Position against Expectation

```python
PLOT = False

full_time = time.time()
# iDate = np.max(global_stats.GAME_DATE) + datetime.timedelta(days=1)
iDate = datetime.datetime.now().date()
# iDate = datetime.date(2016,3,2)
with open('AdjustmentRegressions/opp_vs_pos_%s' % (iDate-datetime.timedelta(days=1)), 'r') as w:
    last = eval(w.read().replace('array','np.array'))
while iDate <= datetime.datetime.now().date():#+datetime.timedelta(days=1):
    iter_time = time.time()
    tab = global_stats[global_stats.GAME_DATE < iDate].copy()
    params = pd.read_csv('Parameters/player_parameters_%s.csv' % iDate, index_col=None).set_index('PERSON_ID')

    print 'Making Rates from Params'
    rates = {}
    for pid in np.unique(tab.Player_ID):
        pid = int(pid)
        try:
            rates[pid] = dfs_utils.params2rates(pd.DataFrame(params.loc[pid]).T)
        except KeyError:
            tab = tab[tab.Player_ID != pid]
    
    N = len(tab.GAME_DATE)
    wgt = np.where(tab.GAME_DATE < datetime.date(2015,10,27), (5./82)*np.ones(N), np.ones(N))
    
    team2regression = {}
    for pos in ('PG','SG','SF','PF','C'):
        sub = tab[tab.POS == pos]
        team2regression[pos] = {}
        for opp in np.unique(sub.OPP):
            if opp not in np.unique(sub[sub.GAME_DATE == iDate - datetime.timedelta(days=1)].OPP):
                team2regression[pos][opp] = last[pos][opp]
                print 'Skipped ', opp
                continue
            sub2 = sub[sub.OPP == opp]
            team2regression[pos][opp] = {}
            
            print 'Making Projections - %s - %s' % (opp, pos)
            st_time = time.time()
            projs = pd.DataFrame()
            for k, row in sub2.iterrows():
                pid = int(row.Player_ID)
                mins = row.MIN
                if mins <= 15:
                    continue
                proj = pd.DataFrame(dfs_utils.rates_mins2simple_projs(rates[pid], mins), columns=[k]).T
                proj.columns = ['PROJ_%s' % x for x in proj.columns]
                projs = projs.append(proj)
            print 'Time Elapsed: ', (time.time() - st_time)
            
            sub3 = sub2.merge(projs, left_index=True, right_index=True)
            for st in ('FGM','FGA','FG3M','FG3A','FTM','FTA','AST','OREB','DREB','BLK','STL','TOV'):
                x = np.array(sub3['PROJ_%s' % st])
                y = np.array(sub3[st])

                coef = reg_lin(x, y, wgt)
                
#                 team2regression[pos][opp][st] = {}
#                 team2regression[pos][opp][st]['func'] = lambda u: reg_eval(u, coef)
                team2regression[pos][opp][st] = np.array([0., coef[0]]) #coef
                
                if PLOT:
                    plt.plot(x, y, 'o', alpha=.2)
                    plt.title(opp + ' - ' + pos + ' - ' + st)
                    plt.xlabel('E_FD')
                    plt.ylabel('FD')
                    plt.grid()
                    
                    X = np.linspace(np.min(x), np.max(x), 200)
                    reg = reg_eval(X, coef)

                    plt.plot(X, reg, 'r*')
                    plt.show()
#             break
#         break
#     break
    with open('AdjustmentRegressions/opp_vs_pos_%s' % iDate, 'w') as w:
        w.write(str(team2regression))
    last = team2regression
    print 'iDate %s Time Elapsed:' % iDate, (time.time() - iter_time)
    iDate += datetime.timedelta(days=1)
print 'Done Time Elapsed:', (time.time() - full_time)
```
