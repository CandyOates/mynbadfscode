```python
%matplotlib inline
import numpy as np
import pandas as pd
import datetime
import statsmodels.api as sm
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,8)

import player as data_pl
import team as data_te
import game as data_ga
```

```python
tid2team = dict(zip([int(data_te.TEAMS[team]['id']) for team in data_te.TEAMS.keys()], data_te.TEAMS.keys()))

table = pd.DataFrame()
for team, dic in data_te.TEAMS.iteritems():
    tid = dic['id']
    gl = data_te.TeamGameLogs(tid, season='2015-16').info().append(data_te.TeamGameLogs(tid, season='2014-15').info(), ignore_index=True)
    table = table.append(gl, ignore_index=True)
table['GAME_DATE'] = [datetime.datetime.strptime(x, '%b %d, %Y').date() for x in table.GAME_DATE]
table
```

```python
iDate = datetime.date(2015, 10, 27)
while iDate <= datetime.datetime.now().date():
    sub_table = table[table.GAME_DATE < iDate]
    tids = np.unique(sub_table.Team_ID).tolist()
    indices = dict(zip(tids, range(30)))

    gids = np.unique(sub_table.Game_ID.tolist())
    ngames = len(gids)

    A = np.zeros((ngames, 30))
    b = np.empty(ngames)
    W = np.eye(ngames, ngames)

    for k, gid in enumerate(gids):
        sub = sub_table[sub_table.Game_ID==gid]
        t1 = sub.iloc[0]
        t2 = sub.iloc[1]
        pos = \
            ((t1.FGA.sum() + 0.4 * t1.FTA.sum() - 1.07 * (t1.OREB.sum() / (t1.OREB.sum() + t2.DREB.sum()))\
            * (t1.FGA.sum() - t1.FGM.sum()) + t1.TOV.sum()) +\
            (t2.FGA.sum() + 0.4 * t2.FTA.sum() - 1.07 * (t2.OREB.sum() / (t2.OREB.sum() + t1.DREB.sum()))\
            * (t2.FGA.sum() - t2.FGM.sum()) + t2.TOV.sum()))

        if t1.GAME_DATE < datetime.date(2015, 7, 1):
            wgt = 5./82
        else:
            wgt = 1.
        W[k,k] = wgt

        mins = sub.MIN.tolist()[0]
        ppm = pos / (mins/5)

        b[k] = ppm
        for tid in sub.Team_ID:
            A[k, indices[tid]] = 1.

    ppms = np.linalg.solve(A.T.dot(W.dot(A)), A.T.dot(W.dot(b)))
    ppms = dict(zip(indices.keys(), [ppms[indices[k]] for k in indices.keys()]))
    pd.DataFrame({'PPM': ppms.values()}, index=ppms.keys()).to_csv('TeamPace/pace_%s' % iDate)
    
    print iDate
    iDate += datetime.timedelta(days=1)
```
